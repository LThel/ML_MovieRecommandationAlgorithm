```python
import pandas as pd
import numpy as np
```

```python
titleakas = pd.read_csv("https://datasets.imdbws.com/title.akas.tsv.gz",sep = "\t",compression = "gzip")
titleakas.head(5)
```

```python
namebasics = pd.read_csv("https://datasets.imdbws.com/name.basics.tsv.gz",sep = "\t",compression = "gzip")
namebasics.head(5)
```

```python
titlebasics = pd.read_csv("https://datasets.imdbws.com/title.basics.tsv.gz",sep = "\t",compression = "gzip")
titlebasics.head(5)
```

```python
titlecrew = pd.read_csv("https://datasets.imdbws.com/title.crew.tsv.gz",sep = "\t",compression = "gzip")
titlecrew.drop('writers', inplace=True, axis=1)
titlecrew.head(5)
```

```python
titleratings = pd.read_csv("https://datasets.imdbws.com/title.ratings.tsv.gz",sep = "\t",compression = "gzip")
titleratings.head(5)
```

```python
titleprincipals = pd.read_csv("https://datasets.imdbws.com/title.principals.tsv.gz",sep = "\t",compression = "gzip")
titleprincipals.head(5)
```

```python
titlebasics = titlebasics[(titlebasics['titleType'] == "movie") | (titlebasics['titleType'] == "tvMovie")]
titlebasics= pd.merge(titlebasics,titleratings,on='tconst',how='left')
titlebasics= pd.merge(titlebasics,titlecrew,on='tconst',how='left')
titlebasics.drop('endYear', inplace=True, axis=1)
titlebasics.to_csv("titles.csv")
```

```python
 namebasics_1stclean = namebasics[namebasics["primaryProfession"].apply(lambda x : ("actor" in x) or ("actress" in x) or ("director" in x) or ("directress" in x) if type(x) == str else False)]
```

```python
namebasics_1stclean.dropna(subset=["knownForTitles"], inplace=True)
namebasics_1stclean["knownForTitles"] = namebasics_1stclean["knownForTitles"].str.split(",")
df = pd.DataFrame({
    "nconst": np.repeat(namebasics_1stclean["nconst"].values, namebasics_1stclean["knownForTitles"].str.len()),
    "knownForTitles": np.concatenate(namebasics_1stclean["knownForTitles"].values)
})
namebasics_2ndclean = pd.merge(df, namebasics_1stclean, on="nconst", how="left")
```

```python
namebasics_2ndclean.rename(columns = {'knownForTitles':'tconst'}, inplace = True)
```

```python
namebasics_2ndclean.drop('knownForTitles_y', inplace=True, axis=1)
namebasics_2ndclean #Table for actors and directors. Must figure out how to get actor names into the main table.
```

```python
namebasics_2ndclean.to_csv("basicsclean.csv")
```
